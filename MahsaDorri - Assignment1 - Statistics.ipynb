```python
df = pd.read_csv('mosquitos_data.csv')
df
```

###    Create side-by-side boxplots for the number of mosquitos in each group (beer vs water) 

```python
fg= plt.figure(figsize=(10,10))
df.boxplot(column="Response",by="Treatment")
plt.xlabel("Treatment")
plt.ylabel("Response")
plt.show()               
```

###    What does the graph reveal about the data for both groups? Is there an association between beer consumption and attractiveness to mosquitos.

As we can see, the Beer group shows a much higher response compared to the Water group. This can be understood by looking at the IQRs of both boxplots.

Beer consumption appears to increase attractiveness to mosquitos.

###    Calculate basic statistics measures for each group: Mean, median, standard deviation and explain the numbers. 

```python
stats_measures = df.groupby("Treatment")["Response"].agg(["mean","median","std", "count"])
stats_measures
```

On average, people who drank beer attracted about 4 to 5 more mosquitos than those who drank water. In other words, mosquitos seemed more drawn to the Beer group.

Looking at the medians, the same story, in the Beer group typically had more mosquitos around them than someone in the Water group.

The spread of responses was pretty similar in both groups (about 3 to 4 mosquitos), so the difference isn’t because of random variation. It really looks like beer made participants more attractive to mosquitos.

### Write code to implement the data simulation demonstrated in the above mentioned video.  Calculate the probability of obtaining the experiment outcome assuming no association between beer consumption.

```python
N=10000
    
results = []
k=0 
df_response= df["Response"].to_numpy()    
for iter in range(N):

    np.random.shuffle(df_response)
    
    mean = np.mean(df_response[:25]) - np.mean(df_response[25:])
    
    results.append(mean)
    
    k += 4.4 < mean
k
```

```python
plt.hist(results, bins=30)
plt.axvline(x=4.4, color='red', linestyle='--')
plt.title('Histogram of Mean Differences')
plt.xlabel('Difference in Means')
plt.ylabel('Frequency')
plt.show()
```
